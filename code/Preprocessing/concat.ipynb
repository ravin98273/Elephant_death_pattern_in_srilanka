```python
import os
import pandas as pd

# Directory path containing the Excel files
directory = r'D:\UOP\4th year\Research\cleaning\Dataset ele death\cleaned'

# List of file names to concatenate
file_names = ['2010_clean', '2011_clean', '2012_clean', '2013_clean', '2014_clean', '2015_clean', '2016_clean', '2017_clean']

# Initialize an empty list to store DataFrames
dfs = []

# Loop through the file names and read each Excel file into a DataFrame
for file_name in file_names:
    file_path = os.path.join(directory, f'{file_name}.csv')
    df = pd.read_csv(file_path)
    dfs.append(df)

# Concatenate all DataFrames into a single DataFrame
concatenated_df = pd.concat(dfs, ignore_index=True)

# Print the concatenated DataFrame
concatenated_df['No'] = concatenated_df.index + 1
concatenated_df
```

```python
output_file_path = r'D:\UOP\4th year\Research\cleaning\Dataset ele death\cleaned\cancat.csv'

concatenated_df.to_csv(output_file_path, index=False)
```
