```python
elephant_deaths_by_ds = df['D/S Office'].value_counts().reset_index()
elephant_deaths_by_ds.columns = ['D/S Office', 'ElephantDeaths']
```

```python
elephant_deaths_by_ds.head(20)
```

```python
merged_data = gdf.merge(elephant_deaths_by_ds, on='D/S Office', how='left')
```

```python
merged_data.head(5)
```

```python
merged_data['ElephantDeaths'].fillna(0, inplace=True)
```

```python
# Set up the plot
fig, ax = plt.subplots(1, 1, figsize=(24, 20))
merged_data.boundary.plot(ax=ax, color='red', linewidth=.1)
# Plot the map 
merged_data.plot(column='ElephantDeaths', cmap='Blues', linewidth=0.8, edgecolor='0.8', legend=True, ax=ax)

# Add district names as labels with an offset above the district boundary
merged_data[merged_data['ElephantDeaths'] > 0].apply(lambda x: ax.annotate(text=x['D/S Office'], xy=x.geometry.centroid.coords[0], ha='center', va='bottom',fontsize=6), axis=1)

# Add the number of deaths as labels with an offset below the district boundary, excluding zeros
merged_data[merged_data['ElephantDeaths'] > 0].apply(lambda x: ax.annotate(text=str(int(x['ElephantDeaths'])), xy=x.geometry.centroid.coords[0], ha='center', va='top'), axis=1)

# Add a title
plt.title('Elephant Deaths by D/S Office in Sri Lanka 2010-2018')

# Show the map
plt.show()
```
