```python
import re
# Function to convert months and weeks to years
def convert_to_years(age_str):
    if isinstance(age_str, str):
        if re.match(r'^\d+(\.\d+)? Months?$', age_str):
            months = float(re.match(r'^\d+(\.\d+)?', age_str).group())
            return months / 12  # Approximate conversion to years
        elif re.match(r'^\d+ Weeks?$', age_str):
            weeks = int(re.match(r'^\d+', age_str).group())
            return weeks / 52  # Approximate conversion to years
        elif re.match(r'^\d+ Days?$', age_str):
            days = int(re.match(r'^\d+', age_str).group())
            return days / 365  # Approximate conversion to years
        elif '-' in age_str:
            age_range = age_str.split('-')
            try:
                return (float(age_range[0]) + float(age_range[1])) / 2  # Taking the average for range
            except ValueError:
                return None  # Invalid range
    try:
        return float(age_str)
    except ValueError:
        return None  # Unknown or invalid value

# Apply the conversion function to the 'Age' column
df['age'] = df['Age'].apply(convert_to_years)

# Function to categorize age
def categorize_age(age):
    if age is None or isinstance(age, str):
        return 'Unknown'
    elif age < 1:
        return 'Infant'
    elif age <= 4:
        return 'Calf'
    elif age <= 12:
        return 'Juvenile'
    elif age >=13:
        return 'Subadults and adults'
    else:
        return 'Unknown'

# Apply the function to the 'Age' column
df['Age Category'] = df['age'].apply(categorize_age)

# Drop the original 'Age' column if you don't need it anymore
df.drop(columns=['age'], inplace=True)

```
